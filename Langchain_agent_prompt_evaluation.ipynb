```
import os
from google.colab import userdata
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.tools import tool
from langchain.agents import create_agent

# Setup Environment and LLM
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
judge_llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# Create prompt template with requirements
eval_prompt_template = ChatPromptTemplate.from_template(
    """
    Evaluate the following user prompt.
    Return Dictionary with score for each category.
    Return a JSON object with the following keys:
    - "Score Category": Catgory based scores
    - "Final_Score": integer between 1-10
    - "Justification": brief explanation
    - "Suggestions": list of 3 suggestions for improvement

    user_prompt: {user_prompt}
    """
)

evaluation_chain = eval_prompt_template | judge_llm


# Define Tool and its function
@tool
def evaluate_prompt(user_prompt: str) -> str:
   """
   Evaluate a user's prompt based on clarity, specificity, context, persona and output constraints.
   Assigns a score 1-10 for each category and then assign final score with an average
   of all category based scores.Finally provides 3 suggestions for improvement.
   """

   result = evaluation_chain.invoke({"user_prompt": user_prompt})
   return result.content

# Create the Agent
agent = create_agent( model=judge_llm, tools=[evaluate_prompt] )

# Invoke Agent and evaluate response
response = agent.invoke({
    "messages": [("user",
        "Evaluate this prompt: How do I choose movies to watch?"
    )]
})

# Read final answer from LLM
output = response["messages"][-1].content
print(output)

```
