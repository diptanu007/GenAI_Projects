```
import os
from langchain_openai import ChatOpenAI
# Retrieving the api key what we just configured
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
llm = ChatOpenAI(model="gpt-4o-mini")
prompt = """You are an executive summary generator.You task is to summarize a 10 page quarterly performance report into an executive summary.Provide concise summary
include key metrics and risks areas.Limit your response in 150 words and provide table metrics and stress on risk and opportunity section of the report.Use facts only and if you
are unable to verify your answer,do not make up response and avoid that area.Use IBM's publicly available Q1 2024 report for quarter results.Provide response in JSON format"""
response = llm.invoke(prompt)
print(response.content)
```

```
import os
from langchain_google_genai import ChatGoogleGenerativeAI


from langchain.agents import create_agent

from langchain_core.tools import tool

from google.colab import userdata
os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')
# Retrieving the api key what we just configured
llm = ChatGoogleGenerativeAI(
     model="gemini-2.5-flash",
     temperature=0 #controls the randomness in the response.
)
```

```
from langchain_core.prompts import ChatPromptTemplate
prompt_template = ChatPromptTemplate.from_messages(
 [
 ("system", "You are a helpful assistant, your goal is to plan a holiday.Please use your tools to plan including travel,stay,recreation activities in the provided {destination} and {number_of_days} for the trip."),
 ("user", "destination:{destination},number_of_days:{number_of_days}")
 ]
)

prompt_value = prompt_template.format_prompt(destination="New York",number_of_days=5)
print(prompt_value.to_messages())
response = llm.invoke(prompt_value)
print(response.content)
```
